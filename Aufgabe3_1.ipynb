```python
# imports pandas library. By using the alias, you can easily access pandas functions and classes
import pandas as pd
```

```python
# this function in the pandas library used to create a DataFrame
doi_df_journal = pd.DataFrame({
    "DOI": [
        "10.1371/journal.pcbi.1004668",
        "10.21105/joss.01035",
        "10.12688/f1000research.18866.2",
        "10.1038/s41598-019-52881-4",
        "10.1186/s12859-019-3171-0"],
    "journal": [
        "PLoS Comput Biol",
        "JOSS",
        "F1000Res",
        "Sci Rep",
        "BMC Bioinformatics"]})

doi_df_first_author = pd.DataFrame({
    "DOI": [
        "10.1371/journal.pcbi.1004668",
        "10.21105/joss.01035",
        "10.21105/joss.01006",
        "10.12688/f1000research.18866.2",
        "10.1186/s12859-019-3171-0"],
    "first_author": [
        "Blischak",
        "Sparks",
        "Granger",
        "Thang",
        "Chen"]})
```

```python
doi_df_first_author
```

```python
doi_df_journal
```

```python
# left: use only keys from left frame, similar to a SQL left outer join; preserve key order
doi_df_first_author.merge(doi_df_journal, on="DOI", how="left")
```

```python
# right: use only keys from right frame, similar to a SQL right outer join; preserve key order
doi_df_first_author.merge(doi_df_journal, on="DOI", how="right")
```

```python
# inner: use intersection of keys from both frames, similar to a SQL inner join; preserve the order of the left keys
doi_df_first_author.merge(doi_df_journal, on="DOI", how="inner")
```

```python
# outer: use union of keys from both frames, similar to a SQL full outer join; sort keys lexicographically
doi_df_first_author.merge(doi_df_journal, on="DOI", how="outer")
```
